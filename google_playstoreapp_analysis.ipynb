Now the values in the column in the Size are string and have some non numeric values in the data
- replace the m with 1000
- replace k with '' empty string
- replace 'Varies with device' with nan
- convert into float
- 

```python
data_copy['Size'] = data_copy['Size'].str.replace('M','000')
```

```python
import numpy as np
data_copy['Size'] = data_copy['Size'].str.replace('Varies with device',str(np.nan))

data_copy['Size'].unique()
```

```python
data_copy['Size'] = data_copy['Size'].str.replace('k','')

data_copy['Size'].unique()
```

```python
data_copy['Size'].dtype
```

```python
data_copy['Size'] = data_copy['Size'].astype('float')
```

```python
data_copy['Size'].dtype
```

```python
data_copy['Size'].head(10)
```

```python
print(19*1000)
print(8.7*1000)
```

To convert the float value i.e. 8.7 into proper 8700

```python
for i in data_copy['Size']:
    if i < 10:
        data_copy['Size'] = data_copy['Size'].replace(i,i*1000)
```

```python
data_copy['Installs']
```

```python
data_copy['Installs'].unique()
```

Here in Installs things to do 
- replace + with '' empty string
- replace , with '' empty string
- convert data type into integer as it cannot be a float value

```python
data_copy['Installs'] = data_copy['Installs'].str.replace(',','').str.replace('+','')
```

```python
data_copy['Installs'] = data_copy['Installs'].astype('int')
```

```python
data_copy['Installs'].head()
```

```python
data_copy.info()
```

```python
char_to_remove = ['+',',','$']
columns_to_clean = ['Price']

for item in char_to_remove:
    for col in columns_to_clean:
        data_copy[col] = data_copy[col].str.replace(item,'')
```

```python
data_copy.head(2)
```

```python
data_copy['Price'].unique()
```

```python
data_copy['Price'] = data_copy['Price'].astype('float')
```

```python
data_copy['Last Updated'].head()
```

```python
data_copy['Last Updated'] = pd.to_datetime(data_copy['Last Updated'])
```
